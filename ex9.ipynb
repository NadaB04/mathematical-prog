```python


pi = 3.141592653589793


### You will need to complete the code for the following two classes
class Point:
    '''This class defines a point in a 2D Euclidean space'''

    def __init__(self, x=0.0, y=0.0):
        '''
        1. The constructor assigns the values of the parameters x and y
        to the respective attributes self.x and self.y.
        '''
        self.x = x
        self.y = y

    def __repr__(self):
        '''
        2. The special method __repr__ returns a string in the form "(x, y)".
        For example, for the point with self.x = 1 and self.y = 2,
        __repr__ returns the string "(1, 2)".
        '''
        return "(" + str(self.x) + ", " + str(self.y) + ")"

    def move(self, dx=0.0, dy=0.0):
        '''
        3. The method move moves the point object by (dx, dy).
        For example, if the Point object p is (1, 1) then p.move(2, 1)
        moves the Point p to (3, 2).
        '''
        self.x = self.x + dx
        self.y = self.y + dy

    def dist(self, other):
        '''
        4. The method dist calculates the Euclidean distance
        between the current Point self and the other Point other.
        The distance d between two points (x1, y1) and (x2, y2) is:
        d = sqrt( (x2 - x1) ** 2 + (y2 - y1) ** 2 )
        For example, if the Point object p1 is (0, 0) and the Point object p2 is (3, 4),
        then p1.dist(p2) returns 5.0.
        '''
        return ((float(other.x) - float(self.x)) ** 2 + (float(other.y) - float(self.y)) ** 2) ** 0.5
        # end of class definition

    pass


class Circle:
    '''This class defines a Circle in a 2D Euclidean space'''

    def __init__(self, point=Point(), R=0.0):
        self.point = point
        if R < 0.0:
            raise ValueError
        else:
            self.R = R


    def area(self):
        return pi * (self.R ** 2)

    def circumference(self):
        return 2 * pi * self.R

    def point_is_on_circumference(self, other):
        if Point.dist(self.point, other) == self.R:
            return True
        else:
            return False

    def contains_point(self, other):
        if Point.dist(self.point, other) < self.R:
            return True
        else:
            return False
    pass




### Don't change the code below.
try:
    x, y = map(float, input("Please enter x,y of center of circle: ").split(","))
    radius = float(input("Please enter the radius of the circle: "))
    my_circle = Circle(Point(x, y), radius)
    print("The area of the circle is:", my_circle.area())
    print("The circumference of the circle is:", my_circle.circumference())
    x, y = map(float, input("Please enter x,y of a point: ").split(","))
    my_point = Point(x, y)
    if my_circle.point_is_on_circumference(my_point):
        print("The point is in the circumference of the circle")
    elif my_circle.contains_point(my_point):
        print("The point is inside the circle")
    else:
        print("The point is outside of the circumference of the circle")

    # The default point is 0,0
    # The defaulat circle center is: 0,0
    # The default circle radius is 0
    assert (Circle().area() == 0)  # area of empty (default) circle is 0!
    assert (not Circle(Point()).contains_point(Point()))  # 0,0 not inside circle with center 0,0 and radius 0!
except ValueError:
    print("There was an input Error!")
```

```python


class InsufficientFundsError(Exception):
    '''an InsufficientFunds Error class to be raised by BankAccount'''
    pass
    
### You will need to complete the code for the following two classes
class BankAccount():
    '''This class defines a Bank Account balance'''
    def __init__(self, change=0):
        '''The constructor creat a new account with balance 0 '''
        self.change = change
    
    def balance(self):
        '''The constructor returns the current value of the account as float number'''
        return float(self.change)
    
    def deposit(self,x):
        '''accepts a positive float x and increases the balance by x'''
        if not x>0 or not isinstance(x,float):
            raise ValueError
        else:
            if not int(x*100)/100==x:
                raise ValueError
            else:
                self.change = self.change + x
    def withdraw(self,x):
        ''' accepts a positive float x and subtracts x from the balance'''
        if not x>0 or not isinstance(x,float):
            raise ValueError
        else:
            if not int(x*100)/100==x:
                raise ValueError
            else:
                if self.change>=x:
                    self.change = self.change - x
                else:
                    raise InsufficientFundsError 
    
    def transfer_from(self, other, x):
        '''transfer funds from other to self.'''
        if isinstance(other, BankAccount):
            self.deposit(x)
            other.withdraw(x)
        else:
            raise ValueError

    def transfer_to(self, other, x):
        '''transfer funds from self to other.'''
        if isinstance(other, BankAccount):
            self.withdraw(x)
            other.deposit(x)
        else:
            raise ValueError

### Don't change the code below.
bank_account1 = BankAccount()
print (bank_account1.balance())
bank_account1.deposit(100.00)
try:
    bank_account1.withdraw(200.0)
except InsufficientFundsError:
    print ("InsufficientFunds")
bank_account1.deposit(301.23)
bank_account1.withdraw(1.23)
print (bank_account1.balance())
```
